```python
#DV2
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
df = sns.load_dataset('titanic')

# Plot a box plot for the distribution of age by sex and survival status
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=df, palette='Set2')

# Adding title and labels
plt.title('Age Distribution by Gender and Survival Status', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.legend(title='Survived', loc='upper left')

# Show the plot
plt.show()
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = sns.load_dataset('iris')

# List of features and types
print(df.dtypes)

# Histograms
df.hist(figsize=(12,8))
plt.suptitle('Histograms of Iris Features')
plt.show()

# Boxplots
plt.figure(figsize=(12,8))
for idx, column in enumerate(df.columns[:-1]):
    plt.subplot(2, 2, idx+1)
    sns.boxplot(y=df[column])
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```
