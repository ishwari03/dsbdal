```python
import seaborn as sns
```

```python
from matplotlib import pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
df
```

**Dist Plot**

```python
plt.figure(figsize=(6,4))
sns.distplot(df['age'])
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.distplot(x=df['fare'],kde=False)
plt.show()
```

```python
sns.jointplot(x='age',y='fare',data=df)
```

```python
sns.jointplot(x='age',y='fare',data=df,kind='hex')
```

```python
plt.figure(figsize=(5,3))
sns.rugplot(x='age',y='fare',data=df)
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax=sns.barplot(x='class',y='age',data=df)
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

```python
plt.figure(figsize=(6,4))
ad=sns.countplot(x=df['class'],hue=df['survived'],palette='rainbow',data=df)
for i in ad.containers:
  ad.bar_label(i)
plt.show()
```

```python
plt.figure(figsize=(6,4))
ad=sns.countplot(x='class',data=df)
for i in ad.containers:
  ad.bar_label(i)
plt.show()
```

```python
plt.figure(figsize=(5,3))
bp= sns.boxplot(x='class',y='age',palette='pastel',data=df)
plt.show()
df.describe().transpose()
```

**Violin Plot**

```python
plt.figure(figsize=(5,4))
vp=sns.violinplot(x='class',y='age',palette='rainbow',data=df)
plt.show()
```

**Strip Plot**

```python
plt.figure(figsize=(6,5))
ab=sns.stripplot(x='class',y='age',data=df)
plt.show()

```

**Hist Plot**

```python
plt.figure(figsize=(5,4))
pq=sns.histplot(x='fare',bins=10,data=df,hue='survived',kde=False)
for i in pq.containers:
  pq.bar_label(i)
plt.show()
```

```python
plt.figure(figsize=(5,4))
st=sns.scatterplot(x='age',y='fare',data=df)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(numeric_only= True), cmap='coolwarm')
plt.show()
```
